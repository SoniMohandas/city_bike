```python
# Dependencies 
import pandas as pd
import numpy as np
```

```python
# City_bike_ride from Jan, 2020 to Dec, 2020 to dataframes
city_bike_01 = pd.read_csv("Data/JC-202001-citibike-tripdata.csv", low_memory=False)
city_bike_01["Month"] = "Janauary"
city_bike_02 = pd.read_csv("Data/JC-202002-citibike-tripdata.csv", low_memory=False)
city_bike_02["Month"] = "February"
city_bike_03 = pd.read_csv("Data/JC-202003-citibike-tripdata.csv", low_memory=False)
city_bike_03["Month"] = "March"
city_bike_04 = pd.read_csv("Data/JC-202004-citibike-tripdata.csv", low_memory=False)
city_bike_04["Month"] = "April"

city_bike_05 = pd.read_csv("Data/JC-202005-citibike-tripdata.csv", low_memory=False)
city_bike_05["Month"] = "May"
city_bike_06 = pd.read_csv("Data/JC-202006-citibike-tripdata.csv", low_memory=False)
city_bike_06["Month"] = "June"
city_bike_07 = pd.read_csv("Data/JC-202007-citibike-tripdata.csv", low_memory=False)
city_bike_07["Month"] = "July"
city_bike_08 = pd.read_csv("Data/JC-202008-citibike-tripdata.csv", low_memory=False)
city_bike_08["Month"] = "August"

city_bike_09 = pd.read_csv("Data/JC-202009-citibike-tripdata.csv", low_memory=False)
city_bike_09["Month"] = "September"
city_bike_10 = pd.read_csv("Data/JC-202010-citibike-tripdata.csv", low_memory=False)
city_bike_10["Month"] = "October"
city_bike_11 = pd.read_csv("Data/JC-202011-citibike-tripdata.csv", low_memory=False)
city_bike_11["Month"] = "November"
city_bike_12 = pd.read_csv("Data/JC-202012-citibike-tripdata.csv", low_memory=False)
city_bike_12["Month"] = "December"
```

```python
# Combining dataframes
df_frame = [city_bike_01, city_bike_02,city_bike_03, city_bike_04, city_bike_05, city_bike_06, 
            city_bike_07, city_bike_08, city_bike_09, city_bike_10, city_bike_11, city_bike_12]
city_bike = pd.concat(df_frame)
city_bike.rename(columns={"start station latitude":"start_lat", "start station longitude":"start_lon", 
                          "end station latitude":"end_lat", "end station longitude":"end_lon", 
                          "start station id":"start_id", "start station name":"start_name", 
                          "end station id":"end_id", "end station name":"end_name"}, inplace=True)
city_bike.head()
```

```python
# Droping null values
city_bike.dropna(inplace = True)

# Reset index
city_bike.reset_index(drop=True, inplace=True)
```

```python
# Import dependency to find the distance between two locations
import haversine as hs
# len(city_bike.starttime.unique())

# Adding distance column which indicates distance convered by each bike_id
def dist(city_bike):
    loc1 = (city_bike["start_lat"], city_bike["start_lon"])
    loc2 = (city_bike["end_lat"], city_bike["end_lon"])
    return hs.haversine(loc1, loc2, hs.Unit.MILES)
city_bike["distance"] = city_bike.apply(dist, axis=1)
city_bike.head()
```

```python
# !pip install uszipcode

# Import packages
from uszipcode import SearchEngine
search = SearchEngine(simple_zipcode=True)
from uszipcode import Zipcode

start_name = [] 
start_zip = {}

#Find start station zipcode
for index, row in city_bike.iterrows():    
    if (row[4] not in start_name):
        start_name.append(row[4])
        s_zip = search.by_coordinates(lat = row[5], lng = row[6], returns = 1)
        start_zip[row[4]] = s_zip[0].zipcode

```

```python
# Find end station zipcode
end_name = []; end_zip={}; i=1
for row in city_bike.itertuples(index=False):    
    if (row[8] not in end_name):
        end_name.append(row[8])
        e_zip = search.by_coordinates(lat = row[9], lng = row[10], returns = 1)
        if e_zip:
            end_zip[row[8]] = e_zip[0].zipcode
            
```

```python
# Add new columns start_zipcode and end_zipcode
city_bike['start_zipcode'] = city_bike['start_name'].map(start_zip)
city_bike['end_zipcode'] = city_bike['end_name'].map(end_zip)
```

```python
# Prepared single dataframe into csv file 
city_bike.to_csv("Data/city_bike.csv", index=False)
```
